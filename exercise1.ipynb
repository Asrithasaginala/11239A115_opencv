```python
from google.colab import files
import cv2
from google.colab.patches import cv2_imshow

uploaded = files.upload()

# Get the uploaded file name
filename = list(uploaded.keys())[0]

# Read the image
img = cv2.imread(filename)

# Display the image
cv2_imshow(img)
```
